```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x, y, z, t, a, b, c, d, e, f, g, h, i, p, q = symbols('x y z t a b c d e f g h i p q')
init_printing()
plt.rcParams['figure.figsize'] = 20, 15
def ltx(a):
  print(latex(a))
```

```python
F = x**3-2*y*z**2
print(F.factor())
P = 1*F.diff(x)+0*F.diff(y)+1*F.diff(z)
P #3x^2-4yz
```

```python
eq = (a*x**2+b*x*y+c*y**2+d*x*z+e*y*z+f*z**2)*(g*x+h*y+i*z)-F
eq = eq.expand()
ltx(eq)
```

```python
ls = []
for pz in range(4):
  for py in range(4):
    for px in range(4):
      if px+py+pz==3:
        ans = eq
        if px>0:
          ans = ans.coeff(x**px)
        if py>0:
          ans = ans.coeff(y**py)
        if pz>0:
          ans = ans.coeff(z**pz)
        ls.append(((px,py,pz), ans))
```

```python
a = Array([r] for _, r in ls)
ltx(a)
a
```

```python
a = Array([r.subs(h,0)] for _, r in ls)
ltx(a)
a
```

```python
t = solve([r for l, r in ls], (a,b,c,d,e,f,g,h,i))
t
```

```python
l = t[0]
((l[0]*x**2+l[1]*x*y+l[2]*x*z+l[3]*y*z+l[4]*z**2)*(l[5]*x+l[6]*y+l[7]*z)).expand()
```

```python
from sympy import *
import numpy as np

x, y, z, w, a, b, c, d, e, f, g, h, i = symbols('x y z w a b c d e f g h i')
init_printing()
```

```python
def f(x, y):
  return (x**2+y**2)**3-4*x**2*y**2
```

```python
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python

xp = []
yp = []

for x in tqdm(np.linspace(-1, 1, 5000)):
  for y in np.linspace(-1, 1, 5000):
    if abs(f(x, y)) < 0.0001:
      xp.append(x)
      yp.append(y)
```

```python
plt.plot(xp, yp, '.')
```

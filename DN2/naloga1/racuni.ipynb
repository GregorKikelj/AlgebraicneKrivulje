```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt

x, y, z, t, a, b, c, d, e, f, g, h, i, p, q = symbols('x y z t a b c d e f g h i p q')
init_printing()
plt.rcParams['figure.figsize'] = 20, 15
```

```python
def T(n) -> Expr:
  if n==0:
    return pi/pi
  elif n==1:
    return t
  else:
    return 2*t*T(n-1) - T(n-2)
```

```python
T(4).expand()
q = (T(6).subs(t, x)-T(4).subs(t, y)).expand()
latex(q.factor())
```

```python
n = 6
p = T(n).subs(t, x) - T(n-1).subs(t, y)
p = p.expand().factor()
p
```

```python
plot_implicit(p)
```

```python
x = []
y = []
px = T(n)
py = T(n-1)

for t0 in np.linspace(-1.01, 1.01, 1000):
  x.append(px.subs(t, t0))
  y.append(py.subs(t, t0))

plt.plot(y, x)
```
